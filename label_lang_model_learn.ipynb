```python
# -*- coding: utf-8 -*-

import tensorflow as tf
import numpy as np
import pandas as pd

from label_lang_model import LabelLangModel
```

```python
train = pd.read_csv('./en_train.csv', sep=',', encoding='utf8')
train['before'] = train.before.astype(unicode)
train['after'] = train.after.astype(unicode)
train.info()
```

```python

sentence_list = []
for (sentence_id, sentence) in train.groupby('sentence_id'):
    sentence_list.append([label for label in sentence['class']])
```

```python

tf.reset_default_graph()

model = LabelLangModel.create_from_data(train['class'].unique())
model.save('./label_lang_model_params.json')


!rm -rf '/tmp/label_lang_model'
sentence_list = np.array(sentence_list)
train_index = int(len(sentence_list) * .8)
model.train('/tmp/label_lang_model', './label_lang_model', 10000, sentence_list[:train_index],
            sentence_list[train_index:], 1000)
```
