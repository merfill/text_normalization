```python
# -*- coding: utf-8 -*-

import tensorflow as tf
import pandas as pd
import numpy as np

from label_classifier import LabelClassifier
```

```python
train = pd.read_csv('./en_train.csv', sep=',', encoding='utf8')
train.dropna(axis=0, how='any', inplace=True)
train['before'] = train.before.astype(unicode)
train['after'] = train.after.astype(unicode)
train.info()
```

```python
train.head()
```

```python

min_class_size = np.min([train[train['class'] == cl].shape[0] for cl in train['class'].unique()]) * 5
result = None
for cl in train['class'].unique():
    t = train[train['class'] == cl][['before', 'class']]
    if t.shape[0] > min_class_size:
        t = t.sample(n=min_class_size)
    if result is None:
        result = t
    else:
        result = pd.concat([result, t])

data = result.values
validation_data = train[['before', 'class']].values
```

```python

tf.reset_default_graph()

model = LabelClassifier.create_from_data(train['before'], train['class'].unique(), state_size=256, num_layers=8)
model.save_params('./label_classifier_256_8_params.json')


!rm -rf '/tmp/lc_classifier'
train_index = int(len(data) * .8)
model.train('/tmp/lc_classifier', './label_classifier_256_8', 10000, data, validation_data, 100)
```
